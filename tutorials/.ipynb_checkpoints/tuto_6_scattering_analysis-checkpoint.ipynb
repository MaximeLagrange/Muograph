<h1><center>Scattering analysis: tutorial</center></h1>


The `analysis` class is used to performed various muon scattering tomography reconstruction algorithm in a convinient way. Tracking and algorithms outputs are saved, allowing to run the algorithms for a large number of muons without memory overload. It has the following arguments:

 - `VOI`: an instance of the `VolumeInterest` class.
 - `input_dir`:`str`, the file where the hits csv file are located.
 - `output_dir`:`str` the directory where to save the analysis outputs.
 - `compute_tracks`:`bool` if true, tracks will be computed from the csv files located in `input_dir`, using the `Tracking` class. If `False`, tracks will be loaded from `output_dir`. Typically, one only has to first run `analysis` with `compute_tracks` = `True`, which will save the tracking data in `output_dir`. Then, simply set `compute_tracks = False` and the tracking data will be loaded.
 - `n_event`:`int` the number of muons to use in the analysis. If `None`, all events will be used.


The following cell shows how to instanciate the analysis class. For a large number of muons (>500k), the tracking is time consuming 

```python
from volume.Volume_Interest import VolumeInterest
VOI = VolumeInterest(position = [0,0,835],dimension=[400,1000,300],voxel_width=20)

from analysis.analysis import analysis
my_analysis = analysis(VOI=VOI,
                       compute_tracks=True,
                       n_event=None,
                       input_dir = "../data/MuoscopeStatue_4Scale_1-2-2016_5M_Bronze_Scattering/G4_data/",
                       output_dir = "../data/MuoscopeStatue_4Scale_1-2-2016_5M_Bronze_Scattering/test_10k/") 
```

```python
from volume.Volume_Interest import VolumeInterest
VOI = VolumeInterest(position = [0,0,835],dimension=[400,1000,300],voxel_width=20)

from analysis.analysis import analysis
my_analysis = analysis(VOI=VOI,
                       compute_tracks=True,
                       n_event=300000,
                       input_dir = "../data/MuoscopeStatue_4Scale_1-2-2016_5M_Bronze_Scattering/G4_data/",
                       output_dir = "../data/MuoscopeStatue_4Scale_1-2-2016_5M_Bronze_Scattering/test_10k/")

my_analysis.poca_setup()
```

```python
# dtheta [Nx,Ny,Nz,[]]
dtheta_voxel_wise = torch.zeros((VOI.n_vox_xyz[0],
                                 VOI.n_vox_xyz[1],
                                 VOI.n_vox_xyz[2])).tolist()

for i in range(VOI.n_vox_xyz[0]):
    for j in range(VOI.n_vox_xyz[1]):
        for k in range(VOI.n_vox_xyz[2]):
            dtheta_voxel_wise[i][j][k]=[]
```

```python
len(dtheta_voxel_wise), len(dtheta_voxel_wise[0]), len(dtheta_voxel_wise[1]) 
```

```python
poca = my_analysis.load_pickle("poca_class")
```

```python
from fastprogress import progress_bar
for ev in progress_bar(range(len(poca.POCA_points))):
    i,j,k = int(poca.indices[ev][0]), int(poca.indices[ev][1]), int(poca.indices[ev][2])
    dtheta_voxel_wise[i][j][k].append(poca.tracks.dtheta[ev])

std_voxel_wise = torch.zeros((VOI.n_vox_xyz[0],
                                 VOI.n_vox_xyz[1],
                                 VOI.n_vox_xyz[2]))

n_poca_per_vox = torch.zeros((VOI.n_vox_xyz[0],
                                 VOI.n_vox_xyz[1],
                                 VOI.n_vox_xyz[2]))

for i in range(VOI.n_vox_xyz[0]):
    for j in range(VOI.n_vox_xyz[1]):
        for k in range(VOI.n_vox_xyz[2]):
            
            n_poca_per_vox[i,j,k] = len(dtheta_voxel_wise[i][j][k])
            
            if(len(dtheta_voxel_wise[i][j][k])>0):
                std_voxel_wise[i,j,k] = np.std(dtheta_voxel_wise[i][j][k])
```

```python
def normalize(tens:torch.tensor) -> torch.tensor:
    
    return (tens - torch.min(tens))/ (torch.max(tens) - torch.min(tens))

from plotting.Plotting import plot_VOI_pred

# plot_VOI_pred(normalize(std_voxel_wise))
plot_VOI_pred(n_poca_per_vox*normalize(std_voxel_wise),filename="../data/weighted")
plot_VOI_pred(n_poca_per_vox,filename="../data/not_weighted")


```
