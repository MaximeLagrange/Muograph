```python
# Usual suspect
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import torch
import sys 
sys.path.insert(1, '../muograph/')

from typing import List

def mean_list(vox_list:List)->float:
    '''
    return the mean of the input list
    '''
    return np.mean(vox_list)

def median_list(vox_list:List, quartile:int)->float:
    '''
    Returns the quartile of the input list.
    Quartile:int must respect 0 < quartile < 100
    '''
    return np.percentile(vox_list,quartile)

def RMS_list(vox_list:List)->float:
    '''
    Return the RMS of the input list.
    '''
    return np.sqrt(np.sum(np.array(vox_list)**2))
```

```python
from volume.Volume_Interest import VolumeInterest

VOI = VolumeInterest(position = [0,0,835],dimension=[400,1000,300],voxel_width=20)

from detector.PixelDetector import PixelDetector
from utils.read_file import hits_csv_to_tensor

hits,E = hits_csv_to_tensor()
det = PixelDetector(hits=hits, xy_min_max = (-500.,500.,-500.,500.),pixel_width=9.,gap=1.)
```

Then we apply the tracking routine.

```python
from tracking.Tracking import Tracking
tracks = Tracking(hits=det.rec_hits,voi=VOI)
```

```python
from reconstruction.ASR import ASR

asr = ASR(tracks=tracks,voi=VOI)
```

```python
score_list = asr.generate_score_list()
score_list,n_true = asr.assign_score_voxel(score_list=score_list,
                               score_feature=tracks.dtheta,
                               masks = [tracks.dtheta>0.01,tracks.dtheta<4])

final_scores = asr.compute_final_voxel_score(score_list=score_list,
                                        score_method=median_list,
                                        quartile=80)

from plotting.Plotting import plot_VOI_pred
plot_VOI_pred(preds=final_scores)
n_true
```
