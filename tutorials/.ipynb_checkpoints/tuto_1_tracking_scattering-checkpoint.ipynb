### Introduction

Tracking is handled by the `Tracking` class of the tracking.py file. Its job is to reconstruct incoming and outgoing muon tracks, compute scattering angle and record informartion relevant to muon scattering tomography image recontruction. 

First we define the Volume of interest using the `VolumeInterest` class.

```python
from volume.Volume_Interest import *

VOI = VolumeInterest(position = [0,0,835],
                     dimension=[600,1000,300],
                     voxel_width=20)
```

```python
from tracking.Tracking import Tracking
from utils.read_file import hits_csv_to_tensor

filename = "../data/Wooden_statue_copper_cylinder/DataCu.csv"
hits,E = hits_csv_to_tensor(n_event=10000,
                            filename = filename)

tracks = Tracking(hits=hits,voi=VOI,E=E,analysis='scattering')
```

The `Tracking` class computes the following features for incoming and outgoing muon:

- `theta`: zenith angle
- `phi`: azimuthal angle
- `tracks`: track direction
- `points`: a point from the track
- `xyz_in_VOI`: muon position when entering the VOI (top of the VOI)
- `xyz_in_VOI`: muon position when exiting the VOI (bottom of the VOI)
- `dtheta`: scattering angle
- `dtheta_x`,`dtheta_y`: projected scattering angle in (XZ and YZ plane)

## Tracking for absorption reconstruction

The Tracking class can also be used in absorption mode, where only the set of detector panels below the VOI are used.

```python
directory = "../data/iron_box_uranium_sphere_absorption/"
filename = "4_planes_abs_iron_box_30cm_from_det_2M_event_2of2.csv"

hits,E = hits_csv_to_tensor(n_event=100000,
                            filename=directory+filename)

VOI = VolumeInterest(position = [0,0,260],
                     dimension=[700,700,500],
                     voxel_width=25)

tracks = Tracking(hits=hits,voi=VOI,E=E,analysis='absorption')
```

```python
from plotting.Plotting import plot_muon_batch
plot_muon_batch(tracks=tracks)
```

```python
from plotting.Plotting import plot_event_reconstruction_abs
plot_event_reconstruction_abs(tracks,hits=True)
```
