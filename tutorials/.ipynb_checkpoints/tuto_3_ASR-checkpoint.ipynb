```python
# Usual suspect
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import torch
import sys
import math

sys.path.insert(1, '../muograph/')

from typing import List

def mean_list(vox_list:List)->float:
    '''
    return the mean of the input list
    '''
    return np.mean(vox_list)

def quartile_list(vox_list:List, quartile:int)->float:
    '''
    Returns the quartile of the input list.
    Quartile:int must respect 0 < quartile < 100
    '''
    return np.percentile(vox_list,quartile)

def rms_list(vox_list:List)->float:
    '''
    Return the RMS of the input list.
    '''
    return np.sqrt(np.sum(np.array(vox_list)**2))
```

<h1><center>Angle Statistic Reconstruction: tutorial</center></h1>

Angle Statistic Reconstruction computation is handled by the `ASR` class of the ASR.py file. The publication describing the working principle can be found [here](https://iopscience.iop.org/article/10.1088/1748-0221/9/11/P11019/meta).

First we define the Volume of interest using the `VolumeInterest` class.

```python
from volume.Volume_Interest import VolumeInterest
VOI = VolumeInterest(position = [0,0,835],
                     dimension=[400,1000,300],
                     voxel_width=20)
```

Then we apply the tracking routine.

```python
from tracking.Tracking import Tracking
from utils.read_file import hits_csv_to_tensor
directory = "../data/Wooden_statue_copper_cylinder/"
filename = "DataCu.csv"
hits,E = hits_csv_to_tensor(n_event=10000,)
tracks = Tracking(hits=hits,voi=VOI,E=E)
```

Now that the tracking is done, one can instanciate the ASR class with the `tracking` and `VOI` instances as arguments.

`ASR` computes the ***triggered* voxels**, the voxels that are **compatible** with both the **incoming** and **outgoing tracks**, and store them in the `triggered_voxels` feature. They are then saved in the `output_dir/ASR/asr_triggered_voxels` file.

```python
from reconstruction.ASR import ASR
asr = ASR(output_dir=directory,tracks=tracks)
```

One can display the event using the `plot_event_reconstruction` function. It takes as arguments:

 - `tracks: Tracks`
 - `VOI : VolumeInterest`
 - `event : Optional[int]`. If provided, display specific event. If not, display random event.
 - `asr: Optional[ASR]`. If provied, display the triggered voxels of the event
 - `hits : bool`. If provided, display the hits.

```python
from plotting.Plotting import plot_event_reconstruction

ev=np.random.randint(1000)

# display a random event and show triggered voxels if any
plot_event_reconstruction(tracks,event=ev,asr=asr)
```

Now that the triggered voxels are computed, we can apply score to them in order to generate the scattering density map. It is done using the following logic:

- 1/ **Assign** an **empty list** $S_{ijk}$ to **each voxel** using `generate_score_list()` method.


- 2/ **for each event**:
    - for each triggered vox in `triggered_vox` with indices $i,j,k$:
        
        - **Append** the desired **feature** `score_feature` (scattering angle $d\theta$ in this case) $S_{ijk}$ **if** the event **satisfies** the **conditions** in `mask` (e.g $d\theta\:>\:0.01\:\text{rad}$ and $d\theta\:<\:1\:\text{rad}$)
        
        
- 3/ **for each voxel**:
    - Assign a final score `final_scores` to the voxel based on its `score_list`, using the desired function `score_method` (`quartile_list` in this case).
    
The `final_scores` tensor can be used to **plot** the **scattering density map** of the VOI, using the `plot_VOI_pred()` method.

The method handling the scattering denisty reconstruction is `asr_analysis` and takes as argument:

 - `VOI:VolumeInterest`, an istance of the VolumeInterest class.
 - `dtheta:torch.tensor`, with size (Nevent) the score_feature used. It should be the muon scattering angle $d\theta$ but other variables could be meaningful (e.g $d\theta^2$).
 - `dtheta_range:List[float]`, the range of scattering angle $\left[d\theta_{min},d\theta_{max}\right]$. Event with $d\theta < d\theta_{min}$ or $d\theta > d\theta_{max}$ will be rejected.
 - `score_method:functools.partial`, the function used for `final_scores` computation.
 - `save:bool`, if `True`, save the final scores in the `output_dir/ASR/method_score_method_**args_dtheta_range[0]_dtheta_range[1]/final_scores` file. 
 - `plot:bool`, if `True`, plot the final scores and save the plots in the `output_dir/ASR/method_score_method_**args_dtheta_range[0]_dtheta_range[1]/` directory.
 
 
It also outputs the final scores.

```python
from functools import partial
final_scores = asr.asr_analysis(VOI = VOI, 
                                dtheta = tracks.dtheta,
                                dtheta_range = [0.,math.pi/2],
                                score_method = partial(np.quantile,q=.5),
                                save=True,
                                plot=True)
```

Another example but using tighter cuts:

- $d\theta\:>\:1\:\text{deg}$
- $d\theta\:<\:7\:\text{deg}$

The choice of the $d\theta$ cuts can act as a filter! Here only remains the copper tube placed inside the wooden statue. Do not hesitate to play with choices of cuts and `score_method`!

```python
from functools import partial
final_scores = asr.asr_analysis(VOI = VOI, 
                                dtheta = tracks.dtheta, #[rad]
                                dtheta_range = [0.0174,0.1221], #[rad]
                                score_method = partial(np.quantile,q=.8),
                                save=True,
                                plot=True)
```
