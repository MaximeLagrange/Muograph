```python
# Usual suspect
import pickle
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import torch
import sys
sys.path.insert(1, '../muograph/')

from typing import List

# Muograph lib
from tracking.Tracking import Tracking
from utils.read_file import hits_csv_to_tensor
from volume.Volume_Interest import *

def mean_list(vox_list:List)->float:
    '''
    return the mean of the input list
    '''
    return np.mean(vox_list)

def quartile_list(vox_list:List, quartile:int)->float:
    '''
    Returns the quartile of the input list.
    Quartile:int must respect 0 < quartile < 100
    '''
    return np.percentile(vox_list,quartile)

def rms_list(vox_list:List)->float:
    '''
    Return the RMS of the input list.
    '''
    return np.sqrt(np.sum(np.array(vox_list)**2))

def sum_list(vox_list:List) -> int:
    """
    Return the sum of the input list
    """
    return np.sum(np.array(vox_list))
```

### B - Angle Statistic Reconstruction

We will use features computed by `tracking` to run a modified version of the ASR algorithm.

**Classical ASR**:

 - **Triggered voxels** are the ones **compatible** with both **incoming and outgoing** tracks.
 - **Apply** a **score** to the triggered **voxels** based on their reconstructed **scattering** angle.
 - Compute voxels **final score** based on **function** provided by the user (mean, RMS, quartile)
 
**Modified ASR**:

 - **Triggered voxels** are the ones **compatible** with the **reconstructed track**.
 - **Apply** a **score** to the triggered **voxels** based on....
 - Compute voxels **final score** based on **function** provided by the user (mean, RMS, quartile)

```python
from reconstruction.absorption import AbsorbtionReconstruction

_abs = AbsorbtionReconstruction(tracks=tracks,voi=VOI)
```

The figure below shows muon detector hits and the triggered voxels for a given event:

```python
# Create an empty list for each voxel
score_list = _abs.generate_score_list()

# Define mask to filter events
masks = None

# Assign score if the event respects the conditions in mask
score_list,n_true = _abs.assign_score_voxel(score_list=score_list,
                                           score_feature=torch.ones_like(tracks.events),
                                           masks = masks)
```

After running the modified ASR, we obtain the figure below. In order to "see the absorbed muons", we must normalize w.r.t the free sky measurement! 

```python
# Compute final voxel scores using the desired function
final_scores = _abs.compute_final_voxel_score(score_list=score_list,
                                             score_method=sum_list,
                                             quartile=None)

directory = "../data/final_scores/"
with open(directory+filename,'wb') as f:
    pickle.dump(final_scores,f)
```

```python
directory = "../data/final_scores/"

with open(directory+filename,'rb') as f:
    final_scores = pickle.load(f)
    
from plotting.Plotting import plot_VOI_pred
plot_VOI_pred(preds=final_scores)
```

## 2 - Free sky measurement



```python
filename = "4_planes_abs_free_sky_30cm_from_det_2M_event_2of2"

# hits,E = hits_csv_to_tensor(filename= filename + ".csv",
#                            directory = "../data/iron_box_uranium_sphere_absorption/")

# tracks = Tracking(hits=hits,voi=VOI,E=E,analysis='absorption')

# directory = "../data/tracking/" 
# tracks.save(filename = directory+filename)
```

```python
directory = "../data/tracking/" 

with open(directory+filename,'rb') as f:
    tracks_free_sky = pickle.load(f)

from reconstruction.absorption import AbsorbtionReconstruction

abs_free_sky = AbsorbtionReconstruction(tracks=tracks_free_sky,voi=VOI)
```

```python
# Create an empty list for each voxel
score_list = abs_free_sky.generate_score_list()

# Define mask to filter events
masks = None

# Assign score if the event respects the conditions in mask
score_list,n_true = abs_free_sky.assign_score_voxel(score_list=score_list,
                                                   score_feature=torch.ones_like(tracks_free_sky.events),
                                                   masks = masks)

# Compute final voxel scores using the desired function
final_scores = abs_free_sky.compute_final_voxel_score(score_list=score_list,
                                             score_method=sum_list,
                                             quartile=None)


directory = "../data/final_scores/"
with open(directory+filename,'wb') as f:
    pickle.dump(final_scores,f)
```

```python
directory = "../data/final_scores/"

with open(directory+filename,'rb') as f:
    final_scores_free_sky = pickle.load(f)
    
from plotting.Plotting import plot_VOI_pred
plot_VOI_pred(preds=final_scores_free_sky)
```

```python
fig,ax = plt.subplots(nrows = 5,ncols=2, figsize=(10,20))
ax = ax.ravel()
from mpl_toolkits.axes_grid1 import make_axes_locatable

for z_plane in range(abs_map.size()[2]):
#     abs_map[:,:,z_plane] = torch.transpose(abs_map[:,:,z_plane],0,1)
#     im = ax[z_plane].imshow(abs_map[:,:,z_plane].numpy())
    im = ax[z_plane].imshow(torch.transpose(abs_map[:,:,z_plane],0,1).numpy())
    
    ax[z_plane].set_title("Voxel z layer = {}".format(z_plane))
    divider = make_axes_locatable(ax[z_plane])
    cax = divider.append_axes("right", size="5%", pad=0.05)
    fig.colorbar(im, ax=ax[z_plane],cax=cax)
plt.tight_layout()
plt.show()
```

```python
plt.rcParams.update({'font.size': 16})

mean = torch.mean(abs_map.ravel())
sig = torch.std(abs_map.ravel())
fig,ax = plt.subplots(figsize=(9,6))
ax.hist(abs_map.ravel(),bins=50,label=r"$\sigma =$ {:.3f}".format(sig))
ax.axvline(x=mean,label='mean = {:.2f}'.format(mean),color='red')
ax.set_ylabel("# voxels per bin")
ax.set_xlabel("transmission")
plt.legend()
plt.show()

abs_map[:,:,0].sum(dim=1)/14
```
